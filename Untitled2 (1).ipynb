```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#1 que
#    a.Scatterplot- Scatter plots are used to display the relationship between two numerical variables. 
#    b.Barplot- Bar plots are effective for comparing values across different categories
#    c.Boxplot- Box plots are used to display the distribution of data and identify potential outliers. 
#    d.Relplot- It displays the relation between two varaibales
#    e.Histogram- Histograms are used to visualize the distribution of a single numerical variable. 
```

```python
#2 que
fmri=sns.load_dataset('fmri')
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',size='region')
```

```python
titanic=sns.load_dataset('titanic')
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='pclass',y='age',data=titanic)

plt.subplot(1,2,2)
sns.boxplot(x='pclass',y='fare',data=titanic)
```

```python
diamond=sns.load_dataset('diamonds')
```

```python
sns.histplot(x='price',hue='cut',data=diamond, multiple='stack', bins=30)
plt.grid()
```

```python
iris=sns.load_dataset('iris')
```

```python
sns.pairplot(data=iris,hue='species')
```

```python
flights=sns.load_dataset('flights')
```

```python
flights_matrix = flights.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap using Seaborn
plt.figure(figsize=(10, 6))  # Set figure size
sns.heatmap(data=flights_matrix, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Heatmap of Flights Dataset')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
